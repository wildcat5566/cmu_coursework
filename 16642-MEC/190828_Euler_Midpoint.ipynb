```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Signal
t = np.linspace(0, 10*np.pi, 501)
#print(t[1]) #0.02*np.pi, 0.0628s

plt.plot(t, np.cos(t))
plt.xlabel('Angle [rad]')
plt.ylabel('cos(t)')
plt.axis('tight')
plt.show()
```
