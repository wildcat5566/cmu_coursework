```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Signal
signal = np.linspace(0, 4, 501) #t[1] = .01
def f(t, y): #d(exp(y))/dy = exp(y)
    return y

T = 1
k = int(4 / T) + 1
t = np.linspace(0, 4, k)

#Euler method
y = [0]*k
y[0] = 1
for i in range(1, k):
    y[i] = y[i-1] + T*f(t[i-1], y[i-1])
    
#Midpoint method
yy = [0]*k
yy[0] = 1
for i in range(1, k):
    yy[i] = yy[i-1] + T*f(t[i-1] + 0.5*T, yy[i-1] + 0.5*T*f(t[i-1], yy[i-1]))
    
#Runge-Kutta method
yyy = [0]*k
yyy[0] = 1
for i in range(1, k):
    k1 = T*f(t[i-1], yyy[i-1])
    k2 = T*f(t[i-1] + 0.5*T, yyy[i-1] + 0.5*k1)
    k3 = T*f(t[i-1] + 0.5*T, yyy[i-1] + 0.5*k2)
    k4 = T*f(t[i-1] + T, yyy[i-1] + k3)
    yyy[i] = yyy[i-1] + (k1 + 2*k2 + 2*k3 + k4)/6

plt.plot(signal, np.exp(signal), 'k') #np.cos(t)
plt.plot(t, y, ':')
plt.plot(t, yy, 'g--')
plt.plot(t, yyy, 'r-.')
plt.xlabel('t')
plt.ylabel('exp(t)')
plt.axis('tight')
plt.show()
```

```python
#Signal
signal = np.linspace(0, 20, 501)

def f(t, y): #d(cos(x)) / dx = -sin(x)
    return -np.sin(t)

T = 1
k = int(20 / T) + 1
t = np.linspace(0, 20, k)

#Euler method
y = [0]*k
y[0] = 1
for i in range(1, k):
    y[i] = y[i-1] + T*f(t[i-1], y[i-1])
    
#Midpoint method
yy = [0]*k
yy[0] = 1
for i in range(1, k):
    yy[i] = yy[i-1] + T*f(t[i-1] + 0.5*T, yy[i-1] + 0.5*T*f(t[i-1], yy[i-1]))
    
#Runge-Kutta method
yyy = [0]*k
yyy[0] = 1
for i in range(1, k):
    k1 = T*f(t[i-1], yyy[i-1])
    k2 = T*f(t[i-1] + 0.5*T, yyy[i-1] + 0.5*k1)
    k3 = T*f(t[i-1] + 0.5*T, yyy[i-1] + 0.5*k2)
    k4 = T*f(t[i-1] + T, yyy[i-1] + k3)
    yyy[i] = yyy[i-1] + (k1 + 2*k2 + 2*k3 + k4)/6

plt.plot(signal, np.cos(signal), 'k')
plt.plot(t, y, ':')
plt.plot(t, yy, 'g:')
plt.plot(t, yyy, 'r:')
plt.xlabel('t')
plt.ylabel('cos(t)')
plt.axis('tight')
plt.show()
```
